#### Program 14
##### Write a Python Program to Check Prime Number.
- A prime number is a whole number that cannot be evenly divided by any other number
except for 1 and itself. For example, 2, 3, 5, 7, 11, and 13 are prime numbers because they
cannot be divided by any other positive integer except for 1 and their own value.

```python
num = int(input("Enter a number: "))
prime = True

if num < 2:
    prime = False
for i in range(2, num):
    if num % i == 0:
        prime = False
        break
print("Prime" if prime else "Not Prime")
```

```python
lower = 1
upper = 10
print("Prime numbers(1-10):")
for num in range(lower,upper + 1):
    # all prime numbers are greater than 1
    if num > 1:
        for i in range(2, num):
             if (num % i) == 0:
                 break
        else:
            print(num)
```

#### Program 16
##### Write a Python Program to Find the Factorial of a Number.

```python
num = int(input('Enter the number'))
fact = 1
if num < 0:
    print('Factorial does not exist for negative numbers')
elif num == 0:
    print('factorial of 0 is 1')
else:
    for i in range(1,num+1):
        fact = fact*i
    print(f'Factorial os {num} is {fact}')
```

#### Program 19
##### Write a Python Program to Check Armstrong Number
- It is a number that is equal to the sum of its own digits, each raised to a power equal to the
number of digits in the number

```python
num = int(input("Enter a number: "))

digits = len(str(num))   # Count number of digits
sum = 0
temp = num

while temp > 0:
    digit = temp % 10          # Get last digit
    sum += digit ** digits     # Raise to power and add
    temp //= 10                # Remove last digit

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
