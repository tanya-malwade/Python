### Program 41
#### Write a Python Program to Remove Punctuation From a String.


```python
string = input("Enter a string: ")
punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
result = ""
for char in string:
    if char not in punctuation:
        result += char
print("String without punctuation:", result)
```

### Program 43
#### Write a Python program to check if the given number is a Disarium Number.
- A Disarium number is a number that is equal to the sum of its digits each raised to the
power of its respective position. For example, 89 is a Disarium number because
8 + = 8 + 81 = 89.

```python
num = int(input("Enter a number: "))

temp = num
length = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** length
    temp //= 10
    length -= 1

if sum == num:
    print(num, "is a Disarium number")
else:
    print(num, "is not a Disarium number")
```

### Program 44
#### Write a Python program to print all disarium numbers between 1 to 100.

```python
for num in range(1, 101):
    temp = num
    length = len(str(num))
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** length
        temp //= 10
        length -= 1

    if total == num:
        print(num, end=" ")
```

### Program 45
#### Write a Python program to check if the given number is Happy Number.
- Happy Number: A Happy Number is a positive integer that, when you repeatedly replace
the number by the sum of the squares of its digits and continue the process, eventually
reaches 1. If the process never reaches 1 but instead loops endlessly in a cycle, the number
is not a Happy Number.


```python
num = int(input("Enter a number: "))

seen = set()

while num != 1 and num not in seen:
    seen.add(num)
    total = 0
    
    while num > 0:
        digit = num % 10
        total += digit ** 2
        num //= 10
    
    num = total

if num == 1:
    print("Happy Number")
else:
    print("Not a Happy Number")
```

### Program 46
#### Write a Python program to print all happy numbers between 1 and 100.

```python
for num in range(1, 101):
    n = num
    seen = []

    while n != 1 and n not in seen:
        seen.append(n)
        total = 0

        while n > 0:
            digit = n % 10
            total += digit * digit
            n //= 10

        n = total

    if n == 1:
        print(num, end=" ")
```

### Program 48
#### Write a Python program to find smallest number in a list.


```python
numbers = [23, 5, 89, 12, 7]

smallest = numbers[0]

for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest number is:", smallest)
```

```python
numbers = [23, 5, 89, 12, 7]

print("Smallest number is:", min(numbers))
```

### Program 49
#### Write a Python program to find second largest number in a list.


```python
numbers = [10, 45, 23, 67, 67, 12]

largest = second_largest = float('-inf')

for num in numbers:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print("Second largest number is:", second_largest)
```

```python
numbers = [10, 45, 23, 67, 67, 12]

numbers = list(set(numbers))  # remove duplicates
numbers.sort()

print("Second largest number is:", numbers[-2])
```

### Program 50
#### Write a Python program to print even numbers in a list.


```python
numbers = [10, 23, 45, 68, 90, 33]

print("Even numbers in the list:")

for num in numbers:
    if num % 2 == 0:
        print(num, end=" ")
```
