#### Program 9
##### Write a Python program to solve quadratic equation.
- ax + bx + c = 0
- The solutions of this quadratic equation is given by: (−b ± (b^2 − 4ac)^1/2)/(2a)

```python
import math
a = float(input("Enter coefficient a:"))
b = float(input("Enter coefficient b:"))
c = float(input("Enter coefficient c:"))

# calculate the descriminant (b^2 - 4ac)
descriminant = b**2 - 4*a*c

# check the descriminant is positive negative or zero
if descriminant > 0 :
    # two real and distinct roots 
    root1 = (-b + math.sqrt(descriminant))/(2*a)
    root2 = (-b - math.sqrt(descriminant))/(2*a)
    print(f"Root 1: {root1}")
    print(f"Root 2: {root2}")
elif descriminant == 0:
    # one real root (repeated)
    root = -b / (2 * a)
    print(f"Root:{root}")
else:
    # complex roots 
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(abs(descriminant)) / (2*a)
    print(f"root 1: {real_part} + {imaginary_part}i")
    print(f"root 2: {real_part} - {imaginary_part}i")
```

#### Program 10
##### Write a Python program to swap two variables without temp variable.

```python
a = float(input("Enter the first number"))
b = float(input("Enter the second number"))

# print before swapping
print(f"Before swapping: a = {a} , b = {b}")

# swap the variables 
a , b = b , a

print(f'After swapping: a = {a} , b = {b}')
```
