#### Program 51
##### Write a Python program to Remove empty List from List.

```python
lists = [[1, 2, 3], [], [4, 5], [], [6, 7, 8], []]

result = []

for item in lists:
    if item != []:
        result.append(item)

print("List after removing empty lists:", result)
```

```python
# Sample list containing lists
list_of_lists = [[1, 2, 3], [], [4, 5], [], [6, 7, 8], []]
# Using a list comprehension to remove empty lists
filtered_list = [i for i in list_of_lists if i]
# Print the filtered list
print("List after removing empty lists:", filtered_list)
```

#### Program 52
##### Write a Python program to Cloning or Copying a list.


```python
# 1. Using Using the Slice Operator
original_list = [1, 2, 3, 4, 5]
cloned_list = original_list[:]
print(cloned_list)
```

```python
# 2. Using the list() constructor
original_list = [1, 2, 3, 4, 5]
cloned_list = list(original_list)
print(cloned_list)
```

#### Program 53
##### Write a Python program to Count occurrences of an element in a list.


```python
numbers = [1, 2, 3, 2, 4, 2, 5]
element = int(input("Enter the element: "))

count = 0

for num in numbers:
    if num == element:
        count += 1

if count == 0:
    print("Element not found")
else:
    print("Occurrences of", element, ":", count)
```

#### Program 54
##### Write a Python program to find words which are greater than given length k.

```python
sentence = input("Enter a sentence: ")
k = int(input("Enter the length k: "))

words = sentence.split()
result = []

for word in words:
    if len(word) > k:
        result.append(word)

print("Words longer than", k, "characters:", result)
```

#### Program 56
##### Write a Python program to split and join a string.

```python
string = input("Enter a sentence: ")

# Split the string into words
words = string.split()

print("After split:", words)

# Join the words with a hyphen
joined_string = "-".join(words)

print("After join:", joined_string)
```

#### Program 57
##### Write a Python program to check if a given string is binary string or not.

```python
string = input("Enter a string: ")

is_binary = True

for char in string:
    if char not in ('0', '1'):
        is_binary = False
        break

if is_binary:
    print("It is a Binary String")
else:
    print("It is NOT a Binary String")
```

#### Program 58
##### Write a Python program to find uncommon words from two Strings.


```python
str1 = input("Enter first sentence: ")
str2 = input("Enter second sentence: ")

words1 = str1.split()
words2 = str2.split()

uncommon = []

for word in words1:
    if word not in words2:
        uncommon.append(word)

for word in words2:
    if word not in words1:
        uncommon.append(word)

print("Uncommon words:", uncommon)
```

#### Program 59
##### Write a Python program to find all duplicate characters in string.

```python
string = input("Enter a string: ")
duplicates = []

for char in string:
    if string.count(char) > 1 and char not in duplicates:
        duplicates.append(char)

print("Duplicate characters:", duplicates)
```
