#### Program 31
##### Write a Python Program for cube sum of first n natural numbers?

```python
num = int(input('Enter the value os n'))
summ = 0
if num<=0:
    print("Enter the no greter than 0")
else:
    for i in range(1,num+1):
        i3 = i**3
        summ = summ + i3
print(f"The cube sum of the first {num} natural numbers is: {summ}")
```

#### Program 33
##### Write a Python Program to find largest element in an array.


```python
arr = [10, 45, 23, 67, 12]

largest = arr[0]

for num in arr:
    if num > largest:
        largest = num

print("Largest element is:", largest)
```

#### Program 34
##### Write a Python Program for array rotation.

```python
arr = [1, 2, 3, 4, 5]

first = arr[0]
for i in range(len(arr) - 1):
    arr[i] = arr[i + 1]

arr[-1] = first
print("Array after left rotation:", arr)
```

#### Program 35

#### Program 36
##### Write a Python Program to check if given array is Monotonic.

```python
arr = [1, 2, 2, 3]

increasing = True
decreasing = True

for i in range(len(arr) - 1):
    if arr[i] > arr[i + 1]:
        increasing = False
    if arr[i] < arr[i + 1]:
        decreasing = False

if increasing or decreasing:
    print("Array is Monotonic")
else:
    print("Array is Not Monotonic")
```

#### Program 37
##### Write a Python Program to Add Two Matrices.


```python
# Matrix 1
A = [
    [1, 2, 3],
    [4, 5, 6]
]

# Matrix 2
B = [
    [7, 8, 9],
    [1, 2, 3]
]

result = []

for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
        row.append(A[i][j] + B[i][j])
    result.append(row)

print("Sum of matrices:")
for r in result:
    print(r)
```

#### Program 38
##### Write a Python Program to Multiply Two Matrices.

```python
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

result = [[0, 0], [0, 0]]

for i in range(len(A)):
    for j in range(len(B[0])):
        for k in range(len(B)):
            result[i][j] += A[i][k] * B[k][j]

print("Matrix Multiplication Result:")
for r in result:
    print(r)
```

#### Program 39
##### Write a Python Program to Transpose a Matrix.


```python
A = [
    [1, 2, 3],
    [4, 5, 6]
]

rows = len(A)
cols = len(A[0])

transpose = []

for j in range(cols):
    row = []
    for i in range(rows):
        row.append(A[i][j])
    transpose.append(row)

print("Transpose of matrix:")
for r in transpose:
    print(r)
```

#### Program 40
##### Write a Python Program to Sort Words in Alphabetic Order.

```python
# Input a sentence from the user
sentence = input("Enter the words: ")
# Split the sentence into words
words = sentence.split()
# Sort the words alphabetically
words.sort()
# Join the sorted words back into a sentence
sorted_sentence = ' '.join(words)
print("Words in alphabetical order:", sorted_sentence)
```
